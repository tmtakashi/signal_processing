## 線形予測分析(LPC)

### 母音音声の生成モデル

$g_0[n]$（周期パルス列） => $G_e(z)$（音源特性） => $V(z)$（声道特性） => $R(z)$（放射特性） => $y[n]$（音声信号）

各部を連結した全システムの伝達特性$H(z)$は

$$
H(z) = \frac{Y(z)}{G_0(z)} = G_e(z)V(z)R(z)\\
Ge(z): -12[\mathrm{db/oct.}]\\
R(z): +6[\mathrm{db/oct.}]
$$

$G_e(z)$と$R(z)$の特性を合わせると-6dB/oct.の特性になる。逆特性+6dB/oct.の特性をもつフィルタによって音源および放射の特性を近似的に取り除く。

$y[n]$ => $\frac{1}{G_e(z)R(z)}$（+6dB/oct.の高域強調フィルタ） => 線形予測分析 => 声道特性$V(z)$

高域強調フィルタ（プリ・エンファシスフィルタ）は
$$
y[n] = x[n] - px[n]
$$
で表せる。$p=0.97$とすることが多い。

### 最小二乗法による線形予測係数の決定

音声データを
$$
y[n], 0 \leq n \leq N - 1
$$
とする。$N$は分析フレームの長さで、通常は20msから30msの長さ。

LPCでは過去のサンプルの線形結合を用いて現時点のサンプルを予測する。
$$
\tilde y[n] = \sum_{k=1}^Pa_ky[n-k]
$$
$P$は予測次数

実際の音声サンプルと予測された音声サンプルの差$e[n]$は

$$
e[n] = y[n] - \tilde y[n]
$$

分析フレーム内の誤差は
$$
C = \sum_{n=0}^{N-1}e[n]^2
$$

これを最小化する条件は
$$
\frac{\partial C}{\partial a_k} = 0   (1 \leq k \leq P)
$$

これを計算すると

$$
\sum_{l=1}^P\phi[l, k]a_l = -\phi[0, k] (1 \leq k \leq P)
$$

ここで$\phi[l, k]$は
$$
\phi[l, k] = \sum_{n=0}^{N-1}y[n-l]y[n-k]
$$

正規方程式の形で表すと
$$
\bf{A}\bf{a} = -\bf{b}
$$

よって、以下の式でパラメータの最適解を求めることができる。
$$
\bf{a} = -A^{-1}\bf{b}
$$

### 前処理、窓処理

```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
```

プリ・エンファシスフィルタを作る

```python
from scipy.signal import lfilter

def pre_emphasis(x, p=0.97):
    b = [1, -p]
    a = [1]
    return lfilter(b, a, x)
```

```python
# 「あ」の音声波形
a_master, fs = sf.read('wavfiles/a_1.wav')
```

```python
fs
```

```python
# 元の音声波形
plt.plot(a_master)
```

```python
# 途中から20ms分取り出す
window_size = int(0.002 * fs)
a_partial = a_master[2000:2000+window_size]

# プリ・エンファシスをかける
a_preemphasized = pre_emphasis(a_partial)

plt.plot(range(len(a_partial)), a_partial, label='before pre-emphasis')
plt.plot(range(len(a_preemphasized)), a_preemphasized, label='after pre-emphasis')
plt.legend()
```

```python
window  = np.hamming(window_size)

a_windowed = a_preemphasized * window

plt.plot(range(len(a_preemphasized)), a_preemphasized, label='before window')
plt.plot(range(len(a_windowed)), a_windowed, label='after window')
plt.legend()
```

```python
# http://aidiary.hatenablog.com/entry/20120415/1334458954

def LevinsonDurbin(r, lpcOrder):
    """Levinson-Durbinのアルゴリズム
    k次のLPC係数からk+1次のLPC係数を再帰的に計算して
    LPC係数を求める"""
    # LPC係数（再帰的に更新される）
    # a[0]は1で固定のためlpcOrder個の係数を得るためには+1が必要
    a = np.zeros(lpcOrder + 1)
    e = np.zeros(lpcOrder + 1)

    # k = 1の場合
    a[0] = 1.0
    a[1] = - r[1] / r[0]
    e[1] = r[0] + r[1] * a[1]
    lam = - r[1] / r[0]

    # kの場合からk+1の場合を再帰的に求める
    for k in range(1, lpcOrder):
        # lambdaを更新
        lam = 0.0
        for j in range(k + 1):
            lam -= a[j] * r[k + 1 - j]
        lam /= e[k]

        # aを更新
        # UとVからaを更新
        U = [1]
        U.extend([a[i] for i in range(1, k + 1)])
        U.append(0)

        V = [0]
        V.extend([a[i] for i in range(k, 0, -1)])
        V.append(1)

        a = np.array(U) + lam * np.array(V)

        # eを更新
        e[k + 1] = e[k] * (1.0 - lam * lam)

    return a, e[-1]
```

```python
P = 20 # 予測次数

r = np.correlate(a_windowed, a_windowed,mode='full')
a, e = LevinsonDurbin(r, P)
```

```python
from scipy.signal import freqz

# 係数を元にフィルタを作る
w, h = freqz(1, a, fs=fs)

H_truth = np.fft.fft(a_partial * window, n=1024)[:1024//2]

plt.plot(w, 20*np.log10(np.abs(h)), label='prediction')
plt.plot(w, 20*np.log10(np.abs(H_truth)), label='ground truth')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
```
